# Análisando mensalidades do Cursos PROUNI 2018
Link do data set https://brasil.io/dataset/cursos-prouni/cursos/

```python
import pandas as pd
import numpy as np
```

```python
#lendo data frame do csv
df = pd.read_csv('cursos-prouni-a37448e1391540bdb72cf6787e381bb0.csv')
df.head()
```

```python
#verifica colunas com itens nulos e mostra a soma de quantos itens nulos tem
df.isnull().sum()
```

```python
#dropando colunas de notas e salvando no data frame
df = df.drop(columns='nota_integral_ampla')
df = df.drop(columns='nota_integral_cotas')
df = df.drop(columns='nota_parcial_ampla')
df = df.drop(columns='nota_parcial_cotas')
```

```python
df.head()
```

```python
#describe dataframe
df.describe()
```

```python
#visualizando média das mensalidades por estado
df.groupby('uf_busca').mensalidade.mean().sort_values()
```

```python
#visualizando mediana das mensalidades por estado
df.groupby('uf_busca').mensalidade.median().sort_values()
```
