```python
x = []
y = []
y2 = []
y3 = []

for i in range(-50,100,5):
    x.append(i)
    y.append(i)
    y2.append(i * 2)
    y3.append(i * 3)
```

```python
plt.scatter(x,y,s=30,c="green",marker="^")
plt.scatter(x,y2,s=30,c="blue",marker="+")
plt.scatter(x,y3,s=30,c="red",marker="o")
```
